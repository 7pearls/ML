We'll work with the Ecommerce Customers csv file from the company. It has Customer info, suchas Email, Address, and their color Avatar. Then it also has numerical value columns:

* Avg. Session Length: Average session of in-store style advice sessions.

## Training and Testing Data

```python
X= customersDF[['Time on App', 'Time on Website', 'Avg. Session Length', 'Length of Membership']]
```

```python
y= customersDF['Yearly Amount Spent']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

## Predicting Test Data

```python
y_predictions = lr.predict(X_test)
```

** Create a scatter plot of the real test values versus the predicted values. **

```python
plot.scatter(y_test, y_predictions, alpha=0.5)
```

## Evaluating the Model

Let's evaluate our model performance by calculating the residual sum of squares and the explained variance score (R^2).

** Calculate the Mean Absolute Error, Mean Squared Error, and the Root Mean Squared Error. **

```python
error = y_test - y_predictions
```

```python
from sklearn import metrics
```

```python
mae = metrics.mean_absolute_error(y_test, y_predictions)
```

```python
mse = metrics.mean_squared_error(y_test, y_predictions)
```

```python
rmse = np.sqrt(mse)
```

```python
print('MAE: ', mae)
print('MSE: ', mse)
print('RMSE: ', rmse)
```
